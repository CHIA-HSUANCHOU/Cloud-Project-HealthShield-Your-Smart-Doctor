```python
import pandas as pd
df = pd.read_csv('ALL_NHANES_MERGED_20072018.csv')
df = df.rename(columns={
    "SEQN": "ID",
    "DIQ010": "diabetes",
    "RIDAGEYR": "age",
    "RIAGENDR":"gender",

    "BMXHT": "height_cm",
    "BMXWT": "weight_kg",
    "BMXBMI": "bmi",
    "BMXWAIST": "waist_cm",

    "BPXSY1":"systolic1",
    "BPXDI1":"diastolic1",
    "BPXSY2":"systolic2",
    "BPXDI2":"diastolic2",
    "BPXSY3":"systolic3",
    "BPXDI3":"diastolic3",

    "LBXGLU":"glucose_fast",
    "LBXIN":"insulin",
    "LBXGH":"HbA1c",
    "LBXTC":"cholesterol_total",
    "LBDHDD":"hdl",
    "LBDLDL":"ldl",
    "LBXTR":"triglycerides",

    "SMQ020": "ever_smoked",
    "MCQ300C": "family_diabetes",
    "ALQ130":"alcohol_drink",
    "PAQ665":"moderate_excercise",
    "PAQ650":"serious_excercise",
    "SLD012":"sleep_time",
    "HUQ010":"health_score"
})
```

```python
df["systolic_avg"] = df[["systolic1", "systolic2", "systolic3"]].mean(axis=1)
df["diastolic_avg"] = df[["diastolic1", "diastolic2", "diastolic3"]].mean(axis=1)
```

```python
import matplotlib.pyplot as plt

continuous_vars = [
    "age",
    "height_cm",
    "weight_kg",
    "bmi",
    "waist_cm",
    "systolic_avg",
    "diastolic_avg",
    "glucose_fast",
    "insulin",
    "HbA1c",
    "cholesterol_total",
    "hdl",
    "ldl",
    "triglycerides",
    "sleep_time",
]

for col in continuous_vars:
    data = df[col].dropna()
    if data.empty:
        continue

    plt.figure()
    plt.hist(data, bins=50)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
categorical_vars = [
    "gender",
    "diabetes",
    "ever_smoked",
    "family_diabetes",
    "alcohol_drink",
    "moderate_excercise",
    "serious_excercise",
    "health_score",
]

for col in categorical_vars:
    counts = df[col].value_counts(dropna=False)

    plt.figure()
    counts.plot(kind="bar")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```
